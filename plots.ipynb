```python
import pickle
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# Load results
with open("all_molecule_results.pkl", "rb") as f:
    all_results = pickle.load(f)

# Choose the molecule to plot
mol_name = "mol_c"
xlim = 20
results_by_init = all_results[mol_name]

# Plot 0–500 epochs for both init strategies
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle(f"VQE Convergence - {mol_name.upper()}", fontsize=16)

for i, (init_name, optim_results) in enumerate(results_by_init.items()):
    ax = axes[i]
    ax.set_title(f"Init: {init_name}")

    for optimizer, result_dict in optim_results.items():
        trace = result_dict["trace"]
        if trace:
            ax.plot(trace[:xlim], label=optimizer)  # Plot only first 501 points

    ax.set_xlim(0, xlim)
    ax.set_xlabel("Epochs")
    if i == 0:
        ax.set_ylabel("Energy")

    ax.grid(True)
    ax.legend(loc="upper right", fontsize='small')
    ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=False, useMathText=False))
    ax.ticklabel_format(style='plain', axis='y')

plt.tight_layout(rect=[0, 0, 1, 0.96])
# plt.savefig(f"vqe_comparison_{mol_name}.png", dpi=300)
plt.show()
```
